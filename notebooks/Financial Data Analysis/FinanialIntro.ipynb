# Price Changes and Financial Returns

```python
close = pd.read_csv("close.csv", index_col = "Date", parse_dates = True)
close
```

```python
msft = close.MSFT.dropna().to_frame().copy()
```

```python
msft
```

```python
msft.rename(columns = {"MSFT":"Price"}, inplace = True) #rename the microsoft column in place
```

```python
msft
```

```python
msft["P_lag1"] = msft.shift(periods = 1) #shift all the prices forward by 1 day
```

### Absolute price changes (difference)

```python
msft["P_diff"] = msft.Price.sub(msft.P_lag1) #Alternative 1
```

```python
msft
```

```python
msft["P_diff2"] = msft.Price.diff(periods = 1) # Alternative 2
```

```python
msft
```

### Relative price changes (Returns)

```python
msft.Price.div(msft.P_lag1) - 1 # Alternative 1
```

```python
msft["Returns"] = msft.Price.pct_change(periods = 1) # Alternative 2
```

```python
msft
```

```python
msft.drop(columns = ["P_lag1", "P_diff", "P_diff2"], inplace = True) # Drop unecessary columns
```

```python
msft
```
