```python
import yfinance as yf
from datetime import datetime, timedelta
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.4f}'.format
plt.style.use("seaborn-v0_8")
```

```python
start = "2014-10-01" #set start date for data we will pull
end = "2023-01-3"    #set end   date for data we will pull
```

```python
symbol = ["BA", "BTC-USD"]
```

```python
df = yf.download(symbol, start, end) #pull data from yahoo finance and place it into a dataframe
#df = df.swaplevel(axis = "columns").sort_index(axis = "columns") #swap inner and outer index levels
df
```

```python
close = df.Close.copy()
#close = df.iloc[:, df.columns.get_level_values(1)=='Close'] #select the close values from a multi-index table or dataframe
close
```

```python
BA = close.BA.dropna().to_frame().copy()
BTC = close["BTC-USD"].dropna().to_frame().copy()
```

```python
BA["Returns"] = BA.BA.pct_change(periods = 1)
BA
```

```python
BTC["Returns"] = BTC["BTC-USD"].pct_change(periods = 1)
BTC
```

```python
BTC.describe()
```

```python
BTCmu = BTC.Returns.mean()
BTCmu
```

```python
BTCsigma = BTC.Returns.std()
BTCsigma
```
