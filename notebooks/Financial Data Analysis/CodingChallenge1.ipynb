```python
import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta
```

```python
start = "2015-01-01" #set start date for data we will pull
end = "2020-12-31"    #set end   date for data we will pull
```

```python
symbol = "GE, BA"
```

```python
df = yf.download(symbol, start, end) #pull data from yahoo finance and place it into a dataframe
```

```python
df
```

```python
norm = df.Close.div(df.Close.iloc[0]) #normalizing close prices for all
norm.dropna().plot(figsize = (15,8), fontsize = 13) #create a plot of all the normalized close prices
plt.legend(fontsize = 13)
plt.show()
```

```python
norm.GE.dropna().plot(figsize = (15,8), fontsize = 13) #create a plot of GE normalized close prices
plt.legend(fontsize = 13)
plt.show()
```
